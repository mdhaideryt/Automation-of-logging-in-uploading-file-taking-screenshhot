```python
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import random,time
from bs4 import BeautifulSoup
import pyautogui

def take_screenshot(driver):
    timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    screenshot_path = os.path.join("C:\\Users\\MD RUMMAN HAIDER\\Downloads", f"screenshot_test2_{timestamp}.png")
    driver.save_screenshot(screenshot_path)

url = "https://demo.dealsdray.com/"
username = "prexo.mis@dealsdray.com"
password = "prexo.mis@dealsdray.com"
file_path = "C:\\Users\\MD RUMMAN HAIDER\\Downloads\\demo-data.xlsx"

driver = webdriver.Chrome()
driver.get(url)
driver.maximize_window()
username_input = driver.find_element(By.NAME, "username")
password_input = driver.find_element(By.NAME, "password")
login_button = driver.find_element(By.XPATH, "//button[contains(text(),'Login')]")
username_input.send_keys(username)
password_input.send_keys(password)
login_button.click()

time.sleep(random.uniform(2.1,4))

order_dropdown = driver.find_element(By.XPATH, "//span[contains(text(),'Order')]")
order_dropdown.click()

orders_button = driver.find_element(By.XPATH, "//span[contains(text(),'Orders')]")
orders_button.click()
time.sleep(random.uniform(1,3))

button=driver.find_element(By.CSS_SELECTOR,'button[class="MuiButton-root MuiButton-contained MuiButton-containedPrimary MuiButton-sizeMedium MuiButton-containedSizeMedium MuiButtonBase-root  css-vwfva9"]')

button.click()
# add_bulk_orders_button = driver.find_element(By.NAME,"Add Bulk Orders")
# add_bulk_orders_button.click()
time.sleep(random.uniform(1,3))
choose_file_button = driver.find_element(By.XPATH, "//input[@type='file']")
choose_file_button.send_keys(file_path)
time.sleep(random.uniform(1,3))
import_button = driver.find_element(By.XPATH, "//button[contains(text(),'Import')]")
import_button.click()
time.sleep(random.uniform(1,3))
validate_button = driver.find_element(By.XPATH, "//button[contains(text(),'Validate Data')]")
validate_button.click()

time.sleep(random.uniform(2.1,4))

#to handle pop-up
pyautogui.press('enter')

#WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//h5[contains(text(),'Uploaded Orders')]")))


take_screenshot(driver)

driver.quit()
```
